```python
import random
from itertools import product
actions = ["Click", "Scroll", "Exit"]
sample_space = list(product(actions, repeat=2))
print("Sample Space S:")
print(sample_space)
print("Total Outcomes:", len(sample_space))
event = [outcome for outcome in sample_space if "Click" in outcome]
theoretical_probability = len(event) / len(sample_space)
print("\nFavorable Outcomes (At least one Click):")
print(event)
print("\nTheoretical Probability of at least one Click:")
print(theoretical_probability)
trials = 1000
count_sum_7 = 0
for _ in range(trials):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    if die1 + die2 == 7:
        count_sum_7 += 1
experimental_probability = count_sum_7 / trials
print("\nExperimental Probability of sum = 7 after 1000 rolls:")
print(experimental_probability)
theoretical_dice_probability = 6 / 36
print("\nTheoretical Probability of sum = 7:")
print(theoretical_dice_probability)
```
