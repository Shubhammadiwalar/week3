```python
import sqlite3
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS mentors (
    mentor_id INTEGER PRIMARY KEY,
    mentor_name TEXT,
    track TEXT
)
""")
mentor_data = [
    (1, "Dr. Mehta", "Data Science"),
    (2, "Ms. Kapoor", "Web Dev"),
    (3, "Mr. Iyer", "AI/ML")
]
cursor.executemany("""
INSERT OR REPLACE INTO mentors VALUES (?, ?, ?)
""", mentor_data)

conn.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS interns (
	intern_id INTEGER PRIMARY KEY,
	intern_name TEXT,
	track TEXT,
	mentor_id INTEGER,
	FOREIGN KEY (mentor_id) REFERENCES mentors(mentor_id)
)
""")
conn.commit()
```

```python
intern_data = [
    (1, "Raj", "Data Science", 1),
    (2, "Priya", "Web Dev", 2),
    (3, "Arjun", "AI/ML", 3)
]
cursor.executemany("""
INSERT OR REPLACE INTO interns VALUES (?, ?, ?, ?)
""", intern_data)
conn.commit()
```

```python
import pandas as pd

query = """
SELECT i.name AS intern_name,
       i.track,
       m.mentor_name
FROM interns i
INNER JOIN mentors m
ON i.track = m.track;
"""
df = pd.read_sql_query(query, conn)
df
```
