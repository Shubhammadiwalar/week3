```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
```

```python
df = pd.read_csv("housing.csv")  # replace with your dataset
```

```python
plt.figure()
sns.scatterplot(x=df["total_rooms"], y=df["median_house_value"])
plt.title("Total Rooms vs Median House Value")
plt.xlabel("Total Rooms")
plt.ylabel("Median House Value")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df["ocean_proximity"], y=df["median_house_value"])
plt.title("Median House Value Distribution by Ocean Proximity")
plt.xticks(rotation=45)
plt.show()
```

```python
correlation = df["total_rooms"].corr(df["median_house_value"])
print("Correlation:", correlation)
```
