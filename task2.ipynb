```python
import random
trials = 100000
success_independent = 0

for _ in range(trials):
    coin = random.choice(["H", "T"])
    die = random.randint(1, 6)
    
    if coin == "H" and die == 6:
        success_independent += 1

print("Independent Probability (Simulation):", success_independent / trials)
print("Theoretical:", 1/12)
success_dependent = 0

for _ in range(trials):
    bag = ["R"] * 5 + ["B"] * 5
    
    first = random.choice(bag)
    bag.remove(first)
    
    second = random.choice(bag)
    
    if first == "R" and second == "R":
        success_dependent += 1

print("Dependent Probability (Simulation):", success_dependent / trials)
print("Theoretical:", 2/9)
```
